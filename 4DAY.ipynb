```python
class Calcalator:
    def setData(self, first, second):    # 멤버 변수에 초기 값을 주기 위한 메서드 
        self.first = first
        self.second = second
    def add(self):
        self.result = self.first + self.second
        return self.result
    def div(self):
        self.result = self.first / self.second
        return self.result
```

```python
calc = Calcalator()
# print(calc.first) 멤버변수가 생략이 되어서 멤버변수에 초기값이 없어 오류 발생

calc.setData(10, 20)     # setData() 라는 메서드를 이용해서 멤버변수에 초기값 주기.
print(calc.first)
print(calc.second)
```

```python
# 객체 생성을 할 때 멤버변수에 초기값을 갖도록

class Calcalator:
    # 객체생성시에 자동으로 실행되는 메서드 : 생성자
    def __init__(self, first, second):    # 멤버 변수에 초기 값을 주기 위한 메서드 
        self.first = first
        self.second = second
    def add(self):
        self.result = self.first + self.second
        return self.result
    def div(self):
        self.result = self.first / self.second
        return self.result
```

```python
calc1 = Calcalator(10, 20)   # 객체 생성
        # 생성자
print(calc1.first)

```

```python
class FourCal(Calcalator):    #상속
    
    def sub(self):
        self.result = self.first - self.second
        return self.result        
    def mul(self):
        self.result = self.first * self.second
        return self.result
```

```python
a4 = FourCal(20, 30)

print(a4.first)
y = a4.add() # 상속 받은거 
print(y)

y = a4.mul() # 클래스에 정의한 메소드
print(y)
```

```python
a5 = FourCal(0, 10)    # 0 / 10   부정 
y = a5.div()
print(y)
```

```python
a5 = FourCal(10, 0)    # 0 / 10   불능
y = a5.div()
print(y)
```

```python

# https://www.weather.go.kr/w/observation/land/city-obs.do

import requests
import pandas     # 데이터 분석
from bs4 import BeautifulSoup

import matplotlib as mpl  # 시각화
import matplotlib.pyplot as plt   # 시각화

# 스크래핑
url = "https://www.weather.go.kr/w/observation/land/city-obs.do"
source = requests.get(url)

# 크롤링
soup = BeautifulSoup(source.content, "html.parser")
# table-col tablesorter tablesorter-blue tablesorterd4396fb0fcf5c
# class="table_header tablesorter-ignoreRow" id="table_header1"

table = soup.find("table", {"class" : "table-col"})
#print(table)

data = []
for tr in table.find_all('tr'):
    ths = list(tr.find_all('th'))
    tds = list(tr.find_all('td'))
    #print(tds)
    for th in ths:
        if th.find('a'):
            point = th.find("a").string
            temp = tds[4].string
            humidity = tds[9].string           
            data.append([point, temp, humidity])

print(data)

# 데이터를 분석
# 데이터 저장

with open("weather.csv", 'w', encoding="utf-8") as f:
    f.write('지역, 온도, 습도 \n')
    for l in data:
        f.write(f"{l[0]},{l[1]},{l[2]} \n")
f.close()


#데이터 처리

df = pandas.read_csv("weather.csv", encoding="utf-8", index_col='지역')
print(df)

city_df = df.loc[['서울', '인천', '부산', '대전']]
print(city_df)

# 시각화
font_name = mpl.font_manager.FontProperties(fname='C:\Windows\Fonts\malgun.ttf').get_name()
mpl.rc('font', family=font_name)

ax = city_df.plot(kind='bar', title='weather', figsize=(12,4))
plt.show()
```

```python
# 유니버설 함수(UFuncs)
l1 = [1,2,3]
l2 = [3,2,1]

x = np.array([1,2,3])
y = np.array([3,2,1])

print(l1 + l2)    # 확장

print(x + y)      # 덧셈 연산
# + 연산자 : np.add() 함수    -> 유니버설 함수
print(np.add(x, y))
```

```python
def add(l1, l2):
    l4 = []
    for idx in range(0, len(l1)):
        l4.append(l1[idx] + l2[idx])
    return l4
```

```python
result = add(l1, l2)
print(result)
```

```python
# 유니버설 함수(UFuncs)
result = x + y
print(result)

result = np.add(x, y)
print(result)

result = x - y
print(result)

result = np.subtract(x,y)
print(result)
result = x * y
print(result)
result = np.multiply(x, y)
print(result)

result = x/y
print(result)

result = np.divide(x,y)
print(result)


result = x % y
print(result)

result = np.mod(x,y)
print(result)


result = x ** y
print(result)

result = np.power(x,y)
print(result)

# numpy = 숫자!!!!!!!!!    숫자 데이터 이므로 숫자 연산자만 유니버셜 함수가 존재한다!!!!




```
